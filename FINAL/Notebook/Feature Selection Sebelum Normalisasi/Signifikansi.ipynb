```python
import numpy as np
from scipy.stats import chi2_contingency

# Confusion Matrix untuk masing-masing model
confusion_matrices = {
    'Tanpa Normalisasi': np.array([[75, 24], [18, 37]]),   # TN, FP, FN, TP
    'Min-Max Scaling': np.array([[84, 15], [14, 41]]),
    'Z-Score (Standard Scaling)': np.array([[83, 16], [16, 39]]),
    'Decimal Scaling': np.array([[77, 22], [19, 36]])
}

# Melakukan uji chi-square untuk setiap confusion matrix
for model, cm in confusion_matrices.items():
    chi2, p, dof, expected = chi2_contingency(cm)
    print(f"Metode Normalisasi: {model}")
    print(f"Chi-Square: {chi2:.4f}")
    print(f"P-value: {p:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies: \n{expected}")
    print("\n" + "-"*50)
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Confusion Matrix untuk masing-masing model
confusion_matrices = {
    'Tanpa Normalisasi': np.array([[75, 24], [18, 37]]),   # TN, FP, FN, TP
    'Min-Max Scaling': np.array([[84, 15], [14, 41]]),
    'Z-Score (Standard Scaling)': np.array([[83, 16], [16, 39]]),
    'Decimal Scaling': np.array([[77, 22], [19, 36]])
}

# Mengambil confusion matrix untuk Tanpa Normalisasi
cm_no_normalization = confusion_matrices['Tanpa Normalisasi']

# Melakukan uji chi-square untuk perbandingan dengan model Tanpa Normalisasi
for model, cm in confusion_matrices.items():
    if model != 'Tanpa Normalisasi':  # Membandingkan dengan model lain
        chi2, p, dof, expected = chi2_contingency([cm_no_normalization, cm])
        print(f"Perbandingan: Tanpa Normalisasi vs {model}")
        print(f"Chi-Square: {chi2:.4f}")
        print(f"P-value: {p:.4f}")
        print(f"Degrees of Freedom: {dof}")
        print(f"Expected Frequencies: \n{expected}")
        print("\n" + "-"*50)
```
