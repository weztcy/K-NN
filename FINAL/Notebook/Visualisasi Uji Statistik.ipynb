# Persiapan Data

```python
import pandas as pd

# Data T-Statistic
data_t = {
    'Metode': [
        'Min-Max Scaling vs Tanpa Normalisasi',
        'Z-Score (Standard Scaling) vs Tanpa Normalisasi',
        'Decimal Scaling vs Tanpa Normalisasi'
    ],
    'Accuracy': [-3.99754, -2.15584, 2.52689],
    'Precision': [-4.12930, -2.21749, 2.73333],
    'Recall': [-3.54493, -2.35725, 2.92463],
    'F1-Score': [-4.02148, -2.25157, 2.77965],
    'Specificity': [-3.60555, -1.78377, 0.91287],
    'ROC AUC': [-14.30525, -2.47726, 0.84161],
}

# Data P-Value
data_p = {
    'Metode': [
        'Min-Max Scaling vs Tanpa Normalisasi',
        'Z-Score (Standard Scaling) vs Tanpa Normalisasi',
        'Decimal Scaling vs Tanpa Normalisasi'
    ],
    'Accuracy': [0.01035, 0.08363, 0.05273],
    'Precision': [0.00909, 0.07737, 0.04112],
    'Recall': [0.01648, 0.06497, 0.03284],
    'F1-Score': [0.01011, 0.07413, 0.03892],
    'Specificity': [0.01545, 0.13454, 0.40318],
    'ROC AUC': [0.00003, 0.05603, 0.43839],
}

# Ubah ke format long
t_stat_df = pd.DataFrame(data_t).melt(id_vars='Metode', var_name='Metrik', value_name='T-Statistic')
p_value_df = pd.DataFrame(data_p).melt(id_vars='Metode', var_name='Metrik', value_name='P-Value')
```

# Visualisasi T-Statistic

```python
import matplotlib.pyplot as plt
import seaborn as sns

custom_palette = {
    'Min-Max Scaling vs Tanpa Normalisasi': 'green',
    'Z-Score (Standard Scaling) vs Tanpa Normalisasi': 'blue',
    'Decimal Scaling vs Tanpa Normalisasi': 'red',
}

plt.figure(figsize=(12, 6))
sns.barplot(data=t_stat_df, x='Metrik', y='T-Statistic', hue='Metode', palette=custom_palette)
plt.axhline(0, color='gray', linestyle='--')
plt.title('Uji T-Statistic Pada Setiap Metrik Evaluasi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Visualisasi P-Value

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=p_value_df, x='Metrik', y='P-Value', hue='Metode', palette=custom_palette)
plt.axhline(0.05, color='red', linestyle='--', label='Threshold 0.05')
plt.title('Uji P-Value Pada Setiap Metrik Evaluasi')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
